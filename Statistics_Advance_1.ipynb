This holds regardless of the underlying distribution of the population, be it normal, Poisson, binomial, or any alternative distribution.

ii.This means the normal distribution serves as the limiting distribution for sample means as the sample size ùëõ becomes large.

**Q2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Define the range of the discrete uniform distribution
low = 1
high = 7  # upper bound is exclusive in scipy.stats.randint

# Create the distribution
dist = randint(low, high)

# Define the values (support)
x = np.arange(low, high)

# Get the PMF values
pmf_values = dist.pmf(x)

# Plot the PMF
plt.bar(x, pmf_values, color='skyblue', edgecolor='black')
plt.title('PMF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.xticks(x)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**Q3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.**

```python
def bernoulli_pmf(x, p):
    """
    Calculate the PMF of a Bernoulli distribution.

    Parameters:
        x (int): The outcome (0 or 1)
        p (float): The probability of success (0 ‚â§ p ‚â§ 1)

    Returns:
        float: The probability P(X = x)
    """
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1 (inclusive).")

    return p if x == 1 else (1 - p)

# Example usage
p = 0.7  # probability of success
print("P(X=0):", bernoulli_pmf(0, p))
print("P(X=1):", bernoulli_pmf(1, p))
```

**Q6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.**

```python


# Parameters of the discrete uniform distribution
a = 1  # start value
b = 6  # end value

# Possible outcomes
x = np.arange(a, b + 1)

# Since it's uniform, each outcome has equal probability
p = 1 / len(x)

# Calculate the CDF by taking cumulative sum of the PMF
pmf = np.full_like(x, p, dtype=float)
cdf = np.cumsum(pmf)

# Plotting the CDF
plt.step(x, cdf, where='post', label='CDF', color='blue')
plt.scatter(x, cdf, color='red')  # mark the CDF points
plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.xticks(x)
plt.ylim(0, 1.1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.show()
```

**Q9.Write a Python function to calculate Z-scores from a dataset and plot them.**

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    """
    Calculates Z-scores from a dataset and plots their histogram.

    Parameters:
        data (list or np.ndarray): The input dataset.

    Returns:
        z_scores (np.ndarray): Array of calculated Z-scores.
    """
    data = np.array(data)
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot histogram of Z-scores
    plt.hist(z_scores, bins=30, density=True, color='skyblue', edgecolor='black')
    plt.title('Histogram of Z-Scores')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.axvline(0, color='red', linestyle='--', label='Mean (Z=0)')
    plt.legend()
    plt.show()

    return z_scores

# Example usage
if __name__ == "__main__":
    # Generate sample data from a normal distribution
    sample_data = np.random.normal(loc=100, scale=15, size=1000)
    z_scores = calculate_and_plot_zscores(sample_data)
```

**Q14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(x, mean, std):
    """
    Calculate the Z-score for a data point.

    Parameters:
        x (float): Data point
        mean (float): Mean of the distribution
        std (float): Standard deviation of the distribution

    Returns:
        float: Z-score
    """
    return (x - mean) / std

def plot_z_score_comparison(z):
    """
    Plot standard normal distribution and mark the Z-score.

    Parameters:
        z (float): Z-score to compare
    """
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    # Plot standard normal curve
    plt.plot(x, y, label='Standard Normal Distribution', color='blue')

    # Fill area under curve up to z
    x_fill = np.linspace(-4, z, 1000)
    y_fill = norm.pdf(x_fill)
    plt.fill_between(x_fill, y_fill, alpha=0.5, color='skyblue', label=f'Area ‚â§ Z ({z:.2f})')

    # Draw vertical line at Z
    plt.axvline(z, color='red', linestyle='--', label=f'Z = {z:.2f}')

    # Labels and styling
    plt.title('Z-score on Standard Normal Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.legend()
    plt.show()

# Example usage
x = 75        # data point
mean = 70     # population mean
std = 10      # population standard deviation

z_score = calculate_z_score(x, mean, std)
print(f"Z-score for x = {x}: {z_score:.2f}")
plot_z_score_comparison(z_score)
```

**Q15.Implement hypothesis testing using Z-statistics for a sample dataset.**

```python
import numpy as np
from scipy.stats import norm

# Given data
mu_0 = 500         # Claimed population mean
sample_mean = 490  # Sample mean
sigma = 20         # Known population standard deviation
n = 50             # Sample size
alpha = 0.05       # Significance level (for two-tailed test)

# Step 1: Calculate the Z-statistic
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Step 2: Calculate the critical Z-value for two-tailed test
z_critical = norm.ppf(1 - alpha / 2)

# Step 3: Calculate the p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Step 4: Make decision
print(f"Z-statistic: {z:.2f}")
print(f"Z-critical (two-tailed, alpha={alpha}): ¬±{z_critical:.2f}")
print(f"P-value: {p_value:.4f}")

if abs(z) > z_critical:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")
```

**Q16. Create a confidence interval for a dataset using Python and interpret the result.**

```python
import numpy as np
from scipy import stats

# Sample dataset (example)
data = [52, 55, 60, 48, 58, 54, 57, 53, 50, 56]

# Step 1: Calculate statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation (ddof=1)
n = len(data)
confidence_level = 0.95
alpha = 1 - confidence_level

# Step 2: Calculate margin of error using t-distribution
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Step 3: Compute confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Print result
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

**Q17.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.**

```python
import numpy as np
from scipy import stats

# Step 1: Generate data from a normal distribution
np.random.seed(42)  # For reproducibility
mu = 100            # True population mean
sigma = 15          # True population standard deviation
n = 50              # Sample size

data = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Calculate sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for sample std dev

# Step 3: Set confidence level and calculate margin of error
confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha / 2, df=n - 1)
margin_of_error = t_critical * (sample_std / np.sqrt(n))

# Step 4: Confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 5: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
import numpy as np

# Parameters for the continuous uniform distribution
a = 10   # lower bound
b = 20   # upper bound
n = 10000  # number of samples

# Step 1: Simulate random variables
samples = np.random.uniform(low=a, high=b, size=n)

# Step 2: Calculate expected value
# Theoretical expected value
expected_theoretical = (a + b) / 2

# Empirical (sample) expected value
expected_empirical = np.mean(samples)

# Output results
print(f"Theoretical Expected Value: {expected_theoretical:.2f}")
print(f"Empirical Expected Value (from {n} samples): {expected_empirical:.2f}")
```

**Q21.Write a Python program to compare the standard deviations of two datasets and visualize the difference.**

```python
import numpy as np
import matplotlib.pyplot as plt

# Example datasets (you can replace these with your own data)
np.random.seed(42)
data1 = np.random.normal(loc=50, scale=5, size=500)   # Dataset 1 with std dev ~5
data2 = np.random.normal(loc=50, scale=15, size=500)  # Dataset 2 with std dev ~15

# Calculate standard deviations
std1 = np.std(data1, ddof=1)
std2 = np.std(data2, ddof=1)

# Plot histograms side-by-side
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

axes[0].hist(data1, bins=30, color='skyblue', edgecolor='black')
axes[0].set_title(f'Dataset 1\nStd Dev = {std1:.2f}')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')

axes[1].hist(data2, bins=30, color='salmon', edgecolor='black')
axes[1].set_title(f'Dataset 2\nStd Dev = {std2:.2f}')
axes[1].set_xlabel('Value')

plt.suptitle('Comparison of Standard Deviations Between Two Datasets')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

**Q22. Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.**

```python
import numpy as np

# Generate data from a normal distribution
np.random.seed(123)
data = np.random.normal(loc=50, scale=10, size=1000)

# Calculate range
data_range = np.max(data) - np.min(data)

# Calculate IQR
q75, q25 = np.percentile(data, [75, 25])
iqr = q75 - q25

# Output results
print(f"Range of the dataset: {data_range:.2f}")
print(f"Interquartile Range (IQR) of the dataset: {iqr:.2f}")
```

**Q23.Implement Z-score normalization on a dataset and visualize its transformation.**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Step 1: Generate a sample dataset (normal but arbitrary mean/std)
np.random.seed(0)
data = np.random.normal(loc=50, scale=10, size=1000)

# Step 2: Apply Z-score normalization
data_normalized = zscore(data)

# Step 3: Plot original and normalized data
fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

# Original data histogram
axes[0].hist(data, bins=30, color='skyblue', edgecolor='black')
axes[0].set_title('Original Data')
axes[0].set_xlabel('Value')
axes[0].set_ylabel('Frequency')

# Normalized data histogram
axes[1].hist(data_normalized, bins=30, color='salmon', edgecolor='black')
axes[1].set_title('Z-score Normalized Data')
axes[1].set_xlabel('Z-score')

plt.suptitle('Effect of Z-score Normalization')
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()

# Optional: Print mean and std before and after normalization
print(f"Original Data Mean: {np.mean(data):.2f}, Std Dev: {np.std(data):.2f}")
print(f"Normalized Data Mean: {np.mean(data_normalized):.2f}, Std Dev: {np.std(data_normalized):.2f}")
```

Q24.Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calc_skewness_kurtosis(data=None, n=1000, mu=0, sigma=1):
    """
    Calculate skewness and kurtosis of a dataset.
    If no data provided, generate normal data with given parameters.

    Parameters:
    - data: array-like, optional. Dataset to analyze.
    - n: int, sample size if data is None.
    - mu: float, mean of normal distribution if data is None.
    - sigma: float, std dev of normal distribution if data is None.

    Returns:
    - skewness: float
    - kurtosis_val: float (excess kurtosis; 0 for normal distribution)
    """
    if data is None:
        np.random.seed(42)
        data = np.random.normal(loc=mu, scale=sigma, size=n)

    skewness = skew(data)
    kurtosis_val = kurtosis(data)  # excess kurtosis (normal dist = 0)

    print(f"Skewness: {skewness:.4f}")
    print(f"Kurtosis (excess): {kurtosis_val:.4f}")

    return skewness, kurtosis_val

# Example usage
calc_skewness_kurtosis()
```
